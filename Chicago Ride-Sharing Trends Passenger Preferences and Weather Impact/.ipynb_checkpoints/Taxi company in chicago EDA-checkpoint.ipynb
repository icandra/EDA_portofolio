```python
# sort top taxi companies by trips amount
top_taxi_comp = taxi_comp_trips.sort_values('trips_amount', ascending=False)


# seperate top 10 taxi
top_10_taxi = top_taxi_comp[:10]


# seperate the rest
other_taxi = top_taxi_comp[10:]
# replace company_name to 'Others'
other_taxi.loc[:, 'company_name'] = 'Others'
# group 'Others'
other_taxi = (other_taxi.groupby('company_name')['trips_amount']
                        .sum()
                        .reset_index())

# join data
top_taxies = (pd.concat([top_10_taxi, other_taxi])
                .sort_values('trips_amount', ascending=False)
                .reset_index(drop=True))

# check the result
top_taxies
```

**Finding:**<br>
- The distribution of trips amount is skewed positively
- The Interquartile Range (IQR) is huge
- The median (Q2) of trips amount is around 178-179 trips and the Q3 is around 2106 trips
- The Flash Cab taxi company gained the highest number of trips in November 2017
- The rest of the top 10 taxies are slightly different from each other

**Insight:**<br>
From 137311 trips that happened in November 2017, the majority of taxi companies (32/64) had a number of trips below 179. Only 16 out of 64 taxi companies could gain more than 2106 trips. Furthermore, the top 10 taxi companies could gain more than 5000 trips. Although the gap between the top taxi companies and the rest is huge, the share trips amount from the top 10 taxi companies seems not to be monopolized by one or several companies.

## Point 3: Testing Hypothesis - Weather and Ride Duration

In this section, We will examine whether the weather conditions influent the ride duration through the `loop_ord_rides` data frame. <br>

Dataframe `loop_ord_rides` contains data on rides from the Loop to O'Hare International Airport (ORD) from November 2017 every Saturday. The table contains the following data: <br>
- `start_ts`: pickup date and time
- `weather_conditions`: weather conditions at the moment the ride started
- `duration_seconds`: ride duration in seconds

Here are the hypotheses that will be tested: <br>
H0 = "The average duration of rides from the Loop to O'Hare International Airport is the same on rainy Saturdays."<br>
H1 = "The average duration of rides from the Loop to O'Hare International Airport changes on rainy Saturdays."<br>

The statistical difference test will present by applying a two-tailed t-test with a confidence level of 95% (alpha level = 0.05). Before that, the test will be preceded by the Levene test to check the variance.

```python
#retrieve the data
loop_ord_rides = pd.read_csv('/datasets/project_sql_result_07.csv')
loop_ord_rides
```

```python
# check data info
loop_ord_rides.info()
```

**Note**:<br>
The data frame contains no null values and there are no suspicious values. <br> 
We will ignore the inappropriate data type in column `start_ts` since the column is not our interest.

```python
# check unique values in weather condition

loop_ord_rides['weather_conditions'].value_counts()
```

```python
# seperate unique values
# define variable `good_weather`

good_weather = loop_ord_rides.query("weather_conditions == 'Good'")
good_weather['weather_conditions'].value_counts()
```

```python
# seperate unique values
# define variable `good_weather`

bad_weather = loop_ord_rides.query("weather_conditions == 'Bad'")
bad_weather['weather_conditions'].value_counts()
```

```python
# check the variance and declare atribute `option_var` in t_test
p_value_levene = st.levene(good_weather['duration_seconds'],
                           bad_weather['duration_seconds']).pvalue


# define alpha value
alpha = 0.05

# define condition for attribute in t_test_ind function
if p_value_levene<alpha:
    option_var=False
else:
    option_var=True
    
    
    
# present two-tailed t-test
# t_test_ind function
p_value=st.ttest_ind(good_weather['duration_seconds'],
                     bad_weather['duration_seconds'],
                     equal_var=option_var).pvalue

if p_value<alpha:
    print('Reject H0 | Accept H1')
else:
    print( "Accept H0 | Reject H1") 
```

```python
print(f"average ride duration in good weather: {good_weather['duration_seconds'].mean()} seconds")
print(f"average ride duration in bad weather: {bad_weather['duration_seconds'].mean()} seconds")
```

**Verdic:**<br>
The average ride duration in the Good weather is significantly different from the Bad weather, resulting in a p-value less than 0.05. The average ride duration in Bad weather is longer than in good weather.<br>

**Insight:**
During the Bad weather, the average ride duration from Loop to O'Hare International Airport could take 7 minutes longer (247s) than in the normal/good weather
